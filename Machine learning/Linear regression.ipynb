```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv("homeprice.csv")
df1
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df1.area,df1.price,color='red',marker='+')
```

```python
new_df = df1.drop('price',axis='columns')
new_df
```

```python
price = df1.price
price
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,price)
```

```python
#predicting home price
reg.predict([[3300]])
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df1.area,df1.price,color='red',marker='+')
plt.plot(df1.area, reg.predict(df1[['area']]),color='blue')
```

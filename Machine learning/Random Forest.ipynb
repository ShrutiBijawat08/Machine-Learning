```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
digits.target_names
```

```python
plt.gray() 
for i in range(4):
    plt.matshow(digits.images[i])
```

```python
# converting data into dataframe

df = pd.DataFrame(digits.data)
df.head()
```

```python
# adding new column target to data frame

df['target']=digits.target
df.head(20)
```

```python
#Train and the model and prediction


X = df.drop('target',axis='columns')
y = df.target
```

```python
y
```

```python
# splitting the data set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
# ensemble is used when we use multiple algorithms like random forest uses multiple decision trees

from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
# we can change model accuracy by changing estimators
#model = RandomForestClassifier(n_estimators=20)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
# confusion matrix

y_predicted = model.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
# apply decision tree classifier & tell me the accuracy
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```

```python
import pandas as pd

df1=pd.read_csv("iris.csv")
df1.head(20)
```

```python
df1.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le_target = LabelEncoder()
df1['Species_n'] = le_target.fit_transform(df1['Species']) # converting species field to numeric values
```

```python
df1
```

```python
df1=df1.drop(['Species'],axis='columns')
df1
```

```python
df1.shape
```

```python
df1[51:100]
```

```python
inputs=df1.drop(['Species_n'], axis='columns')
target=df1['Species_n']
```

```python
target
```

```python
target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train,y_test= train_test_split(inputs, target, test_size=0.3)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=40)
model.fit(x_train, y_train)
```

```python
model.score(x_test,y_test)
```

```python
model = RandomForestClassifier(n_estimators=10)
model.fit(x_train, y_train)
model.score(x_test,y_test)
```

```python
model = RandomForestClassifier(n_estimators=40)
model.fit(x_train, y_train)
model.score(x_test,y_test)
```

```python
model.predict(x_test)
```

```python
y_test
```

```python
# using confusion matrix

y_predicted = model.predict(x_test)
from sklearn.metrics import confusion_matrix

# we are supplying actual value and predicted value
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import seaborn as sn
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
#plt.show()
```
